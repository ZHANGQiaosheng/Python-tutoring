# 🎯 小学生学Python：打印函数 print() 🎯

欢迎来到Python的神奇世界！今天我们要学习一个超级有用的功能：**print()函数**！

想象一下，print()就像是电脑的"嘴巴"，它可以把我们想说的话显示在屏幕上！ 🗣️

## 🌟 第一步：学会让电脑说话

让我们开始最简单的打印！就像学说话一样，我们先教电脑说"你好"！

```python
# 让电脑说第一句话！
print("你好，世界！")
print("我是小明，今年10岁！")

# 让电脑说出你的名字（记得把"小红"改成你的名字）
print("我的名字是小红")

# 电脑也会数学！
print(5 + 3)
print(10 - 2)

# 电脑还认识True和False（真和假）
print(True)
print(False)
```

## 🎭 第二步：让电脑一次说很多话

有时候我们想让电脑一次说好几样东西，就像这样：

```python
# 电脑一次说很多话（用逗号分开）
print("我喜欢", "苹果", "香蕉", "橘子")
print("我的年龄是", 10, "岁")

# 看！电脑在每个词之间自动加了空格！

# 如果我们不想要空格怎么办？
print("红", "黄", "蓝", sep="")  # sep=""表示中间什么都不要

# 或者我们想用其他符号分开
print("苹果", "香蕉", "橘子", sep=" | ")  # 用 | 分开
print("2024", "9", "25", sep="-")  # 用 - 分开，变成日期
```

## 📦 第三步：给信息起个名字（变量）

有时候我们想把信息装在"盒子"里，这样就不用每次都重新写了！这些"盒子"叫做**变量**。

```python
# 把我的信息装在"盒子"里
我的名字 = "小华"
我的年龄 = 9
我的身高 = 130.5
我喜欢的颜色 = "蓝色"

# 现在让电脑说出盒子里的内容
print("名字:", 我的名字)
print("年龄:", 我的年龄)
print("身高:", 我的身高, "厘米")
print("喜欢的颜色:", 我喜欢的颜色)

# 我们还可以用英文做变量名
name = "小明"
age = 8
print("My name is", name)
print("I am", age, "years old")
```

## ✨ 第四步：神奇的f-string（超级厉害的打印方法！）

现在我要教你一个**超级超级厉害**的方法！它叫f-string，就像变魔术一样！

只要在字符串前面加个小写字母 `f`，然后用花括号 `{}` 包住变量名，就能直接把变量放进句子里！

```python
# 设置我们的信息
姓名 = "小丽"
年龄 = 10
最喜欢的动物 = "小猫"

# 🎩 魔法开始！看f-string有多神奇！
print(f"大家好！我的名字是{姓名}")
print(f"我今年{年龄}岁了")
print(f"我最喜欢的动物是{最喜欢的动物}")

# 更厉害的是，我们还能在里面做计算！
print(f"我明年就{年龄 + 1}岁了！")
print(f"我的名字有{len(姓名)}个字")

# 比较一下普通方法和f-string
print("普通方法：", "我叫", 姓名, "，今年", 年龄, "岁")
print(f"f-string：我叫{姓名}，今年{年龄}岁")

# 看！f-string是不是更简单更漂亮？
```

## 🎨 第五步：让数字变得更好看

有时候数字太长了，我们想让它们看起来更整齐。f-string可以帮我们做到！

```python
# 圆周率有很多很多小数
π = 3.14159265359

# 让我们只看2位小数
print(f"π大约等于 {π:.2f}")

# 或者看4位小数
print(f"π大约等于 {π:.4f}")

# 我的身高和体重
身高 = 135.6789
体重 = 32.1234

print(f"我的身高是 {身高:.1f} 厘米")  # 只要1位小数
print(f"我的体重是 {体重:.0f} 公斤")    # 不要小数

# 大数字也能变好看
很大的数字 = 1234567
print(f"这个数字是：{很大的数字}")
print(f"加逗号更好看：{很大的数字:,}")

# 百分比也很简单！
我的考试成绩 = 0.95  # 95%
print(f"我考了 {我的考试成绩:.0%}！")
```

## 🌈 第六步：做一个漂亮的自我介绍

让我们用学到的知识做一个超棒的自我介绍吧！

```python
# 填写你自己的信息（记得改成你的真实信息哦！）
我的名字 = "小强"
我的年龄 = 9
我的班级 = "三年级2班"
我的爱好 = "画画"
我最喜欢的科目 = "数学"
我的考试成绩 = 92.5

# 🎉 超级自我介绍开始！
print("=" * 30)
print("🌟 我的自我介绍 🌟")
print("=" * 30)
print()

print(f"👋 大家好！我叫{我的名字}")
print(f"🎂 我今年{我的年龄}岁了")
print(f"🏫 我在{我的班级}学习")
print(f"❤️  我最喜欢{我的爱好}")
print(f"📚 我最拿手的科目是{我最喜欢的科目}")
print(f"📝 我上次考试得了{我的考试成绩:.1f}分")

# 计算一些有趣的事情
print()
print("🔢 一些有趣的数字：")
print(f"   我明年就{我的年龄 + 1}岁了")
print(f"   我的名字有{len(我的名字)}个字")
print(f"   我出生在{2024 - 我的年龄}年")

print()
print("🎈 谢谢大家！让我们做好朋友吧！")
print("=" * 30)
```

## 🎮 第七步：做个小游戏 - 数学小助手

让我们用print()和f-string做一个有趣的数学小助手！

```python
# 🧮 数学小助手
print("🎯 欢迎来到数学小助手！")
print()

# 选择两个数字
数字1 = 15
数字2 = 8

print(f"今天我们要算：{数字1} 和 {数字2}")
print()

# 加法
加法答案 = 数字1 + 数字2
print(f"➕ {数字1} + {数字2} = {加法答案}")

# 减法
减法答案 = 数字1 - 数字2
print(f"➖ {数字1} - {数字2} = {减法答案}")

# 乘法
乘法答案 = 数字1 * 数字2
print(f"✖️  {数字1} × {数字2} = {乘法答案}")

# 除法
除法答案 = 数字1 / 数字2
print(f"➗ {数字1} ÷ {数字2} = {除法答案:.2f}")

print()
print("🎉 数学真有趣！")

# 🏆 挑战：看看谁是最大的数？
我的成绩 = [95, 87, 92, 88, 96]
最高分 = max(我的成绩)
最低分 = min(我的成绩)
总分 = sum(我的成绩)
平均分 = 总分 / len(我的成绩)

print()
print("📊 我的成绩分析：")
print(f"   最高分：{最高分}分 🏆")
print(f"   最低分：{最低分}分")
print(f"   总分：{总分}分")
print(f"   平均分：{平均分:.1f}分")
```

## 🕒 第八步：时间魔法师

让我们学会用电脑显示今天的日期和时间！这就像是一个时间魔法师！

```python
# 导入时间魔法包
from datetime import datetime

# 获取现在的时间
现在 = datetime.now()

print("⏰ 时间魔法师来啦！")
print()

# 显示完整的时间
print(f"现在的时间是：{现在}")
print()

# 让我们把时间分开显示
年 = 现在.year
月 = 现在.month
日 = 现在.day
小时 = 现在.hour
分钟 = 现在.minute

print("📅 今天是：")
print(f"   年：{年}年")
print(f"   月：{月}月")  
print(f"   日：{日}日")

print()
print("🕐 现在的时间：")
print(f"   {小时}点{分钟}分")

print()
# 更漂亮的显示方式
print(f"📆 今天是 {年}年{月}月{日}日")
print(f"🕐 现在是 {小时:02d}:{分钟:02d}")

# 星期几呢？
星期名字 = ["星期一", "星期二", "星期三", "星期四", "星期五", "星期六", "星期日"]
今天是星期几 = 现在.weekday()
print(f"🌟 今天是{星期名字[今天是星期几]}")

print()
print("🎈 时间过得真快呀！")
```
